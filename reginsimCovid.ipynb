```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import  train_test_split
from sklearn.linear_model import LinearRegression


dataset = pd.read_csv('db.muertes_hist.csv')

dataset.head(5)
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values

xt = dataset[{'muertes'}]

x
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=0)

x_train

x_test
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
viz_train = plt
viz_train.scatter(x_train, y_train, color = 'blue')
viz_train.plot(x_train, regressor.predict(x_train), color = 'black')
viz_train.title('Muertes vs Edad(Entrenamiento)')
viz_train.xlabel('Edad')
viz_train.ylabel('Muertes')
viz_train.show()
```

```python
viz_test = plt
viz_test.scatter(x_test, y_test, color = 'red')
viz_test.plot(x_train, regressor.predict(x_train), color = 'black')
viz_test.title('Muertes vs Edad(Prediccion)')
viz_test.xlabel('Edad')
viz_test.ylabel('Muertes')
viz_test.show()
```
